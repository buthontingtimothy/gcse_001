```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
dict_param = {
'param_seemark_zh_TW' : {'url': 'https://seemark.hk/seemarkdirectory', 'page_param': '_page', 'start': 1, 'end': 8, 'extract_method': "a[class='_self cvplbd']"},
'param_seemark_zh_CN' : {'url': 'https://seemark.hk/zh_CN/seemarkdirectory', 'page_param': '_page', 'start': 1, 'end': 8, 'extract_method': "a[class='_self cvplbd']"},
'param_seemark_en_GB' : {'url': 'https://seemark.hk/en_GB/seemarkdirectory', 'page_param': '_page', 'start': 1, 'end': 8, 'extract_method': "a[class='_self cvplbd']"},
'param_company_zh' : {'url': 'https://sedirectory.org.hk/zh-hant/sedb', 'page_param': 'page', 'start': 0, 'end': 54, 'extract_method': "a[href^='/zh-hant/sedb/ses/']:not([class='ses-list--se--logo'])"},
'param_company_en' : {'url': 'https://sedirectory.org.hk/en/sedb', 'page_param': 'page', 'start': 0, 'end': 54, 'extract_method': "a[href^='/en/sedb/ses/']:not([class='ses-list--se--logo'])"}
}
```

```python
def get_list(url=None, page_param='page', start=None, end=None, extract_method=None, sorted=True):
    results = []
    for page in range(start, end + 1):    
        response = requests.get(f"{url}?{page_param}={page}")
        soup = BeautifulSoup(response.text, "html.parser")
        results.extend([tab.text.strip() for tab in soup.select(extract_method)])
    if sorted:
        results.sort()
    return results
```

```python
list_combine = []
for param in dict_param:
    list_combine.extend(get_list(**dict_param[param]))
```

```python
list_combine_unique = list(set(list_combine))
```

```python
list_combine_unique
```

```python
df = pd.DataFrame(list_combine_unique, columns=['name']).sort_values('name').reset_index(drop=True)
```
